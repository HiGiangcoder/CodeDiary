# Cách vẽ Histogram và KDE bằng seaborn

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Tạo dữ liệu ngẫu nhiên
data = np.random.rand(1000)  # 1000 giá trị từ phân phối chuẩn

# Vẽ biểu đồ histogram và KDE
sns.histplot(data, bins=30, kde=True, color='blue')

# Thêm tiêu đề và nhãn trục
plt.title('Histogram & KDE')
plt.xlabel('Giá trị')
plt.ylabel('Tần suất')

# Hiển thị biểu đồ
plt.show()
```

### Cách vẽ nhiều biểu đồ trên cùng một figure (subplot)

```python
data2 = (data - data.mean()) / data.std()  # Sau BatchNorm (chuẩn hóa)

# Tạo figure với 2 subplot (1 hàng, 2 cột)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Vẽ histogram + KDE trên subplot đầu tiên
sns.histplot(data, bins=30, kde=True, color='blue', ax=axes[0])
axes[0].set_title('Trước BatchNorm')

# Vẽ histogram + KDE trên subplot thứ hai
sns.histplot(data2, bins=30, kde=True, color='red', ax=axes[1])
axes[1].set_title('Sau BatchNorm')

# Căn chỉnh khoảng cách giữa các subplot
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```
